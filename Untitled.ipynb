```python
# ======================================================
# IMPORTS
# ======================================================
import pandas as pd
import numpy as np
import re
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
RANDOM_STATE = 123
```

```python
# ======================================================
# LOAD DATA
# ======================================================
df = pd.read_csv("Reviews.csv")  


df = df[['Text', 'Score']].dropna()

# ======================================================
# CREATE SENTIMENT LABEL
# ======================================================
def map_score_to_sentiment(score):
    if score <= 2:
        return 'negative'
    elif score == 3:
        return 'neutral'
    else:  # 4-5
        return 'positive'

df['sentiment'] = df['Score'].apply(map_score_to_sentiment)
```

```python
# ======================================================
# TEXT CLEANING FUNCTION
# ======================================================
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", "", text)       
    text = re.sub(r"[^a-zA-Z\s]", " ", text)         
    text = re.sub(r"\s+", " ", text).strip()         
    return text

df['clean_text'] = df['Text'].apply(clean_text)
```

```python
# ======================================================
# TRAIN-TEST SPLIT
# ======================================================
X_train, X_test, y_train, y_test = train_test_split(
    df['clean_text'],
    df['sentiment'],
    test_size=0.2,
    random_state=RANDOM_STATE,
    stratify=df['sentiment']
)

print("Train distribution:\n", y_train.value_counts())
print("Test distribution:\n", y_test.value_counts())
```

```python
# ======================================================
# PIPELINE & GRID SEARCH - NAIVE BAYES
# ======================================================
nb_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=2000, ngram_range=(1,1), min_df=3, stop_words='english')),
    ('nb', MultinomialNB())
])

nb_params = {
    'nb__alpha': [0.1, 0.5, 1.0]
}

nb_grid = GridSearchCV(nb_pipeline, nb_params, cv=5, scoring='accuracy', n_jobs=-1)
nb_grid.fit(X_train, y_train)
y_pred_nb = nb_grid.predict(X_test)

print("\n=== NAIVE BAYES ===")
print("Best hyperparameters:", nb_grid.best_params_)
print(classification_report(y_test, y_pred_nb))
print("Accuracy:", accuracy_score(y_test, y_pred_nb))
```

```python
# ======================================================
# PIPELINE & GRID SEARCH - LOGISTIC REGRESSION
# ======================================================
lr_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=2000, ngram_range=(1,1), min_df=3, stop_words='english')),
    ('lr', LogisticRegression(max_iter=1000, solver='liblinear', random_state=RANDOM_STATE))
])

lr_params = {
    'lr__C': [0.01, 0.1, 1.0, 10.0],
    'lr__penalty': ['l2']
}

lr_grid = GridSearchCV(lr_pipeline, lr_params, cv=5, scoring='accuracy', n_jobs=-1)
lr_grid.fit(X_train, y_train)
y_pred_lr = lr_grid.predict(X_test)

print("\n=== LOGISTIC REGRESSION ===")
print("Best hyperparameters:", lr_grid.best_params_)
print(classification_report(y_test, y_pred_lr))
print("Accuracy:", accuracy_score(y_test, y_pred_lr))

# ======================================================
# CONFUSION MATRICES
# ======================================================
print("\nNaive Bayes Confusion Matrix:\n", confusion_matrix(y_test, y_pred_nb))
print("\nLogistic Regression Confusion Matrix:\n", confusion_matrix(y_test, y_pred_lr))
```

```python

cm_nb = confusion_matrix(y_test, y_pred_nb, labels=['negative', 'neutral', 'positive'])
cm_lr = confusion_matrix(y_test, y_pred_lr, labels=['negative', 'neutral', 'positive'])

disp_nb = ConfusionMatrixDisplay(confusion_matrix=cm_nb, display_labels=['negative', 'neutral', 'positive'])
disp_nb.plot(cmap=plt.cm.Blues, values_format='d')
plt.title("Naive Bayes Confusion Matrix")
plt.show()

disp_lr = ConfusionMatrixDisplay(confusion_matrix=cm_lr, display_labels=['negative', 'neutral', 'positive'])
disp_lr.plot(cmap=plt.cm.Greens, values_format='d')
plt.title("Logistic Regression Confusion Matrix")
plt.show()

```

Handle imbalanced data

```python
# undersampling
# ======================================================
train_data = pd.DataFrame({'review_text': X_train, 'sentiment': y_train})


min_count = train_data['sentiment'].value_counts().min()

balanced_train = train_data.groupby('sentiment', group_keys=False).apply(
    lambda x: x.sample(n=min_count, random_state=RANDOM_STATE)
).reset_index(drop=True)

X_train_bal = balanced_train['review_text']
y_train_bal = balanced_train['sentiment']
```

```python
# ======================================================
# Naive Bayes 
# ======================================================

nb_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=2000, ngram_range=(1,1), min_df=3, stop_words='english')),
    ('nb', MultinomialNB())
])

nb_params = {
    'nb__alpha': [0.1, 0.5, 1.0]
}

nb_grid = GridSearchCV(nb_pipeline, param_grid=nb_params, cv=5, scoring='accuracy', n_jobs=-1)
nb_grid.fit(X_train_bal, y_train_bal)

y_pred_nb = nb_grid.predict(X_test)

print("=== NAIVE BAYES ===")
print("Best hyperparameters:", nb_grid.best_params_)
print(classification_report(y_test, y_pred_nb))
print("Accuracy:", accuracy_score(y_test, y_pred_nb))


cm_nb = confusion_matrix(y_test, y_pred_nb, labels=['negative','neutral','positive'])
print("Naive Bayes Confusion Matrix (raw counts):\n", cm_nb)
disp_nb = ConfusionMatrixDisplay(cm_nb, display_labels=['negative','neutral','positive'])
disp_nb.plot(cmap=plt.cm.Blues, values_format='.2f')
plt.title("Naive Bayes Confusion Matrix")
plt.show()
```

```python
# ======================================================
# Logistic Regression 
# ======================================================
lr_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=2000, ngram_range=(1,1), min_df=3, stop_words='english')),
    ('lr', LogisticRegression(max_iter=1000, solver='liblinear', random_state=RANDOM_STATE))
])

lr_params = {
    'lr__C': [0.01, 0.1, 1.0, 10.0],
    'lr__penalty': ['l2'],
    'lr__class_weight': [None, 'balanced'] 
}

lr_grid = GridSearchCV(lr_pipeline, param_grid=lr_params, cv=5, scoring='accuracy', n_jobs=-1)
lr_grid.fit(X_train_bal, y_train_bal)

y_pred_lr = lr_grid.predict(X_test)

print("\n=== LOGISTIC REGRESSION ===")
print("Best hyperparameters:", lr_grid.best_params_)
print(classification_report(y_test, y_pred_lr))
print("Accuracy:", accuracy_score(y_test, y_pred_lr))


cm_lr = confusion_matrix(y_test, y_pred_lr, labels=['negative','neutral','positive'])
print("Logistic Regression Confusion Matrix (raw counts):\n", cm_lr)
disp_lr = ConfusionMatrixDisplay(cm_lr, display_labels=['negative','neutral','positive'])
disp_lr.plot(cmap=plt.cm.Greens, values_format='.2f')
plt.title("Logistic Regression Confusion Matrix")
plt.show()
```

```python
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import roc_curve, auc, RocCurveDisplay


def plot_multi_roc_macro(models_dict, X_test, y_test):

    lb = LabelBinarizer()
    y_test_bin = lb.fit_transform(y_test)

    plt.figure(figsize=(10,7))
    
    linestyles = {'NB_raw':'--', 'NB_bal':'--', 'LR_raw':'-', 'LR_bal':'-'}
    colors = {'NB_raw':'blue', 'NB_bal':'cyan', 'LR_raw':'green', 'LR_bal':'lime'}

    for name, model in models_dict.items():
        y_score = model.predict_proba(X_test)
        
        for i, class_label in enumerate(lb.classes_):
            fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score[:, i])
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, linestyle=linestyles[name], color=colors[name], alpha=0.3,
                     label=f'{name} {class_label} (AUC={roc_auc:.2f})')

        # macro-average
        all_fpr = np.unique(np.concatenate([roc_curve(y_test_bin[:, i], y_score[:, i])[0] 
                                            for i in range(y_test_bin.shape[1])]))
        mean_tpr = np.zeros_like(all_fpr)
        for i in range(y_test_bin.shape[1]):
            fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_score[:, i])
            mean_tpr += np.interp(all_fpr, fpr, tpr)
        mean_tpr /= y_test_bin.shape[1]
        macro_auc = auc(all_fpr, mean_tpr)
        plt.plot(all_fpr, mean_tpr, color=colors[name], linestyle=linestyles[name], lw=2,
                 label=f'{name} macro-AUC={macro_auc:.2f}')


    plt.plot([0,1], [0,1], color='gray', lw=1, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curves with Macro-Average: Original vs Balanced Training')
    plt.legend(loc='lower right', fontsize=12)

    plt.tight_layout()

    plt.savefig('roc_macro.pdf', format='pdf', dpi=300)
    plt.show()
```

```python
models_dict = {
    'NB_raw': nb_grid_raw,
    'NB_bal': nb_grid_bal,
    'LR_raw': lr_grid_raw,
    'LR_bal': lr_grid_bal
}

plot_multi_roc_macro(models_dict, X_test, y_test)
```
